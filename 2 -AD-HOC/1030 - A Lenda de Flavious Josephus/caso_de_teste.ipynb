```python
def kill(grupo:list, mortes_da_rodada:list):
    for morte in mortes_da_rodada:
        if len(grupo) != 1:
            grupo.remove(morte)
```

```python
n = 1234
k = 223
pessoas = list(range(1,n+1))
cnt = 0
mortes = []
kaux = k
while True:
    print(cnt, len(pessoas), kaux, k)
    
    if kaux <= len(pessoas):
        cnt += kaux
    else:
        kaux = k % len(pessoas)
        cnt += kaux
    
    if cnt >= len(pessoas):
        cnt -= len(pessoas)
        if mortes:
            kill(pessoas, mortes)
            mortes = []
            
    if len(pessoas) == 1:
        break
    
    if cnt == 0:
        cnt = 1
        
    mortes.append(pessoas[cnt-1])
```
